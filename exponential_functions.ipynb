```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
```

```python
display(Math('\\text{Exponential function } f(x) = 2^x'))
```

```python
a,b,x = sym.symbols('a,b,x')
# intial_value
a = 1
# growth_factor
b = 2
```

```python
expr = a*(b**x)
```

```python
x_points = np.linspace(-10, 10)
y_points = []

for x_ in x_points:
    y_ = sym.Float((expr.subs({ 'x': x_ })))
    y_points.append(y_)
```

```python
# y_points
```

```python
fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(12,10))

ax0.plot(x_points, y_points, alpha=0.5, marker='o')
# plot horizontal asymptote
ax0.plot(x_points, np.zeros(len(x_points)), '--', color='orange', alpha=0.8)

plt.show()
```

```python
display(Math('\\text{Exponential function } f(x) = -4^{2-x} + 3'))
```

```python
a,b,x = sym.symbols('a,b,x')
# intial_value
a = -1
# growth_factor
b = 4
expr = a*(b**(2-x))
```

```python
x_points = np.linspace(-5, 5)
y_points = []

for x_ in x_points:
    y_ = sym.Float((expr.subs({ 'x': x_ })))
    y_points.append(y_)
```

```python
fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(12,10))

ax0.plot(x_points, y_points, alpha=0.5, marker='o')
# plot horizontal asymptote
ax0.plot(x_points, (np.zeros(len(x_points)) + 3), '--', color='orange', alpha=0.8)

plt.show()
```

```python
display(Math('\\text{Exponential function } x = -(\\frac{1}{2})^{y-6} + 1'))
```

```python
x,y = sym.symbols('x,y')
```

```python
expr = -1 * ( (1/2) ** (y - 6) ) + 1
```

```python
# expr.subs({ y: 100 })
display(Math('\\text{With }y=%g:\ \Rightarrow \\quad %g' %(100,expr.subs({ y: 100 }))))
```

```python
# display(expr.subs({ y: -100 }))  
display(Math('\\text{With }y=%g:\ \Rightarrow \\quad %g' %(-100,expr.subs({ y: -100 }))))
```

```python
y_points = np.linspace(-100, 100)
x_points = []

for y_ in y_points:
    x_ = expr.subs({ 'y': y_ })
    x_points.append(x_)
```

```python
fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(12,10))

ax0.plot(x_points, y_points, alpha=0.5, marker='o')
# plot vertical asymptote
ax0.plot(np.zeros(len(y_points)) + 1, y_points, '--', color='orange', alpha=0.8)

plt.show()
```

```python
x,y = sym.symbols('x,y')
expr = 3 ** (y+2)
```

```python
# expr.subs({ y: 100 })
display(Math('\\text{With }y=%g:\ \Rightarrow \\quad %g' %(100,expr.subs({ y: 100 }))))
# display(expr.subs({ y: -100 }))  
display(Math('\\text{With }y=%g:\ \Rightarrow \\quad %g' %(-100,expr.subs({ y: -100 }))))
```
