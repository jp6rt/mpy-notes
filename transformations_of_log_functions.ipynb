```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
```

```python
y_points = np.linspace(-5, 5)
```

```python
display(Math('\\text{Log function } f(x)=log_2(x)'))
```

```python
def vshift(d=0):
    x_points = []
    for y in y_points:
        x = 2 ** (y+d)
        x_points.append(x)
    
    return [x_points, y_points]

fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(10,6))

[x0, y0] = vshift()
ax0.plot(x0, y0, alpha=0.5, marker='.', label="log_2(x)")

[x1, y1] = vshift(-2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="log_2(x)+2")

[x2, y2] = vshift(2)
ax0.plot(x2, y2, alpha=0.5, marker='.', label="log_2(x)-2")

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.legend()
plt.show()
```

```python
def hshift(d=0):
    x_points = []
    for y in y_points:
        x = 2 ** (y) + d
        x_points.append(x)
    
    return [x_points, y_points]

fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(10,6))

[x0, y0] = hshift()
ax0.plot(x0, y0, alpha=0.5, marker='.', label="log_2(x)")

[x1, y1] = hshift(-2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="log_2(x+2)")

[x2, y2] = hshift(2)
ax0.plot(x2, y2, alpha=0.5, marker='.', label="log_2(x-2)")

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.legend()
plt.show()
```

```python
display(Math('\\text{Log function } log_2(x)'))
display(Math('\\text{x-axis reflection } -log_2(x)'))
display(Math('\\text{y-axis reflection } log_2(-x)'))
```

```python
def xy_reflect(a=1, c=1):
    x_points = []
    for y in y_points:
        x = a * (2 ** (c * y))
        x_points.append(x)
    
    return [x_points, y_points]

fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(10,6))

[x0, y0] = xy_reflect()
ax0.plot(x0, y0, alpha=0.5, marker='.', label="log_2(x)")

[x1, y1] = xy_reflect(c=-1)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="-log_2(x)")

[x2, y2] = xy_reflect(a=-1)
ax0.plot(x2, y2, alpha=0.5, marker='.', label="log_2(-x)")

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.legend()
plt.show()
```

```python
def v_stretch_compress(c=1):
    x_points = []
    for y in np.linspace(-2,2):
        x = 2 ** (y * c)
        x_points.append(x)
    
    return [x_points, y_points]

fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(10,6))

[x0, y0] = v_stretch_compress()
ax0.plot(x0, y0, alpha=0.5, marker='.', label="log_2(x)")

[x1, y1] = v_stretch_compress(2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="2 log_2(x)")

[x1, y1] = v_stretch_compress(1/2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="1/2 log_2(x)")

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.legend()
plt.show()
```

```python
def h_stretch_compress(a=1):
    x_points = []
    for y in np.linspace(-2,2):
        x = a * (2 ** y)
        x_points.append(x)
    
    return [x_points, y_points]

fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(10,6))

[x0, y0] = h_stretch_compress()
ax0.plot(x0, y0, alpha=0.5, marker='.', label="log_2(x)")

[x1, y1] = h_stretch_compress(2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="log_2(2x)")

[x1, y1] = h_stretch_compress(1/2)
ax0.plot(x1, y1, alpha=0.5, marker='.', label="log_2(1/2 x)")

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.legend()
plt.show()
```
