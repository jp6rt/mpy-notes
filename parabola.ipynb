```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math
```

```python
display(Math('\\text{Parabola stadard form } y = ax^2-bx+c'))
```

```python
x,y,x_coff = sym.symbols('x, y, x_coff')
x_coff = -8
expr = x**2 + x_coff*x + 13

display(Math('\\text{Graph the parabola } y = x^2-8x+13'))
```

```python
# Complete the square to get the vertex
# The vertex form equals to y = (x-4)^2 + 13

b = (x_coff/2)**2

vertex_x = sym.solve(x**2 + x_coff*x + int(b))[0]
# 
vertex_y = 13 - int(b)
```

```python
fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(5,5))

# This plot creates fixed limit on x,y line
line_p = 15
line_x = np.linspace(-(line_p), line_p, 100)
ax0.plot(line_x, np.sin(line_x*np.pi), alpha=0.05, color='grey')
ax0.plot(np.sin(line_x*np.pi), line_x, alpha=0.05, color='grey')

# vertex = 4,-3

x_axis = []
y_axis = []

# Create plots starting from the value of x=0 and end in symmetry with the first plot
for i in range(0, vertex_x*2+1):
    x_axis.append(i)
    l = int(expr.subs({ x : i }))
    y_axis.append(l)

ax0.plot(x_axis, y_axis, marker='o')
ax0.plot(vertex_x, vertex_y, marker='d', color="red")

# plot the axis of symmetry
x_axis_symmetry = []
y_axis_symmetry = []

for i in range(min(y_axis)-5, max(y_axis)+5):
    x_axis_symmetry.append(vertex_x)
    y_axis_symmetry.append(i)
    
ax0.plot(x_axis_symmetry, y_axis_symmetry, '--', color='orange', alpha=0.5)

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.plot(1, 0, ">k", transform=ax0.get_yaxis_transform(), clip_on=False)
ax0.plot(0, 1, "^k", transform=ax0.get_xaxis_transform(), clip_on=False)

plt.show()
```
