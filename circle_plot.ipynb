```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display,Math

plt.style.use('default')
```

```python
fig, (ax0) = plt.subplots(nrows=1, layout='constrained', figsize=(5,5))

# This plot creates fixed limit on x,y line
line_p = 10
x = np.linspace(-(line_p), line_p, 100)
ax0.plot(x, np.sin(x*np.pi), alpha=0.01, color='grey')
ax0.plot(np.sin(x*np.pi), x, alpha=0.01, color='grey')

center_point = (-1, -2)
radius = 4

cir = plt.Circle(center_point, radius, color='r',fill=False)
ax0.add_patch(cir)
ax0.plot(center_point[0], center_point[1], marker='o')

ax0.spines[["left", "bottom"]].set_position(("data", 0))
ax0.spines[["top", "right"]].set_visible(False)

ax0.plot(1, 0, ">k", transform=ax0.get_yaxis_transform(), clip_on=False)
ax0.plot(0, 1, "^k", transform=ax0.get_xaxis_transform(), clip_on=False)

plt.show()
```
